#### Chart - 10

````python
# Chart - 10 visualization code
plt.rcParams['figure.figsize'] = (20, 5)
fig, axs = plt.subplots(ncols=3)
sns.scatterplot(x="Total day calls", y="Total eve calls", hue="Churn", data=df,ax=axs[0])
sns.scatterplot(x="Total day calls", y="Total night calls", hue="Churn", data=df,ax=axs[1])
sns.scatterplot(x="Total night calls", y="Total eve calls", hue="Churn", data=df,ax=axs[2])
#Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

This specific chart is a set of three scatterplots that show the relationship between the number of calls made during the day, evening, and night and the churn rate (whether customers leave or stay) in a telecom company.

The chart is useful for visualizing the relationship between these variables and comparing them side by side. 

Overall, this chart provides an effective way to explore the relationship between the number of calls and the churn rate in our telecom data set.

##### 2. What is/are the insight(s) found from the chart?

Based on the scatterplots, it appears that the churned and non-churned customers are randomly distributed with respect to the number of calls made during the day, evening, and night.

##### 3. Will the gained insights help creating a positive business impact? 
Are there any insights that lead to negative growth? Justify with specific reason.

Although the scatterplots did not reveal any specific reasons why some customers stopped using the phone service, but the insights we gained can still be used to identify other factors that may be causing churn.

For example, we can analyze how much the service costs, how good it is, and whether people like it compared to other services. 

By improving these factors, we can reduce churn and retain more customers, which will help the business grow.
